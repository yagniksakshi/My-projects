# Worldometer Table(Population by country)

```python
from bs4 import BeautifulSoup as bs
import requests #module
```

```python
#this is formal reuest to particular web site to allow scrapping data from this site 
url="https://www.worldometers.info/world-population/population-by-country/"
r=requests.get(url)
r
```

```python
#this code is extract html code(content) of table
htmlcontent=r.content
htmlcontent
```

```python
#tthe below code is to simplyfy the output of above and make it easily readable.(prettify)
Soup=bs(htmlcontent,"html.parser")
print(Soup.prettify())
```

```python
table = Soup.find("table", id="example2")  # table id=example2 #get headers of table

# Check if the table was found
if table is not None:
    # Extract the table headers
    headers = table.find_all("th")

    # Iterate over the headers and extract the text
    header_texts = [header.text.strip() for header in headers]
    print(header_texts)
else:
    print("Table not found on the webpage.")
```

```python
#In this code we will extract table data from html code.(headers data

)
table=Soup.find("table", id="example2") #exaple
if table is not None:
    # Extract the table rows
    rows = table.find_all("tr") #tr=table row
    for row in rows:
        cells = row.find_all("td")#td=table data
        row_data = [cell.text.strip() for cell in cells]#sritp()=remove all spaces at begning and end
        print(row_data)
else:
    print("Table not found on the webpage.")
```
