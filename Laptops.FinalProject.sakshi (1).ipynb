# Laptop's Price Dataset

About Dataset-
This dataset provides a comprehensive collection of information on various laptops, enabling a detailed analysis of their specifications and pricing. It encompasses a wide range of laptops, encompassing diverse brands, models, and configurations, making it a valuable resource for researchers, data analysts, and machine learning enthusiasts interested in the laptop industry.


Fields included:

Laptop Name: The unique identifier or model name of the laptop.
Brand: Laptop brand.
Model: Laptop brand model.
CPU (Central Processing Unit): The processor brand, model, and other relevant details.
GPU (Graphics Processing Unit): The graphics card brand, model, and associated specifications.
RAM (Random Access Memory): The amount of memory available for multitasking.
Storage: The storage type (HDD, SSD) and capacity of the laptop.
Price: The cost of the laptop in the respective currency.
By utilizing this dataset, researchers and analysts can explore patterns, trends, and relationships between laptop specifications and their pricing. It serves as an excellent resource for tasks such as price prediction, market analysis, and comparison of different laptop configurations.

```python
#Histogram Using Metplotlip
plt.figure(figsize=(7,7))
plt.hist(df['Final Price'],bins=10) #histogram univarient he single col ko pakad k krte h to y axis ki value dalne k liye bins=10 k diffrence diya h
plt.title("Price Frequency Distribution")
plt.xlabel("Final Price")
plt.ylabel("Frequency") #width ki value kitni bar ari h vo y axis h
plt.grid(True)
plt.show()
```

```python
#Plot Storage mean and median
P_mean=df['Storage'].mean()
P_median=df['Storage'].median()

plt.figure(figsize=(6,6))
df['Storage'].hist(bins=10)
plt.title("Storage Frequency Distribution")#steps fix rhege fig and axvline vale
plt.axvline(P_mean,color='red',label='mean') #this axvline is use to drow mean and median line(line se identify hota h)
plt.axvline(P_median,color='yellow',label='median')
plt.legend()
plt.xlabel('Storage')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar Graph using metplotlip
plt.figure(figsize=(8,8))
df['Brand'].value_counts().plot(kind='bar')
plt.title("Laptop Brands count bar plot")
plt.xlabel("Brand")
plt.ylabel("count")
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```

```python
#Bar Graph using seaborn
plt.figure(figsize=(7,7))
sns.countplot(data=df,x='Brand')
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar Graph using seaborn plot Brand with respect to RAM
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='Brand',hue='RAM')
plt.xticks(rotation=90)
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(7,7))

df['Brand'].value_counts().plot.pie()
plt.title("Laptop's Brand Contribution Data",color='green')
plt.show()
```
